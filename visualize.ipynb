```python
# prac9.1.1
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('company_sales_data.csv')
month = data['month_number']
profit = data['total_profit']
plt.plot(month,profit)
plt.xlabel('month')
plt.ylabel('profit')
plt.xticks(month)
plt.grid()
plt.title('Profit Over Months')
plt.show()
```
